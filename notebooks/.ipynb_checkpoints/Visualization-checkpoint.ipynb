```python
df_votes = pd.read_csv('../data/votes.csv')
df_votes.head(10)
```

```python
for i,row in enumerate(df_votes.iterrows()):
    if row[1].votes_gop_2016>row[1].votes_dem_2016:
        df_votes.loc[i,'16Winner'] = 'Trump'
    else:
        df_votes.loc[i,'16Winner'] = 'Clinton'  
    if row[1].votes_gop_2012>row[1].votes_dem_2012:
        df_votes.loc[i,'12Winner'] = 'Romney'
    else:
        df_votes.loc[i,'12Winner'] = 'Obama' 
        
df_votes.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df_votes['16Winner'])
df_votes['16Winner'] = le.transform(df_votes['16Winner'])
le.fit(df_votes['12Winner'])
df_votes['12Winner'] = le.transform(df_votes['12Winner'])

df_votes.head(10)
```

2016 Winners:
    0: Clinton
    1: Trump
    
2012 Winners:
    0: Obama
    1: Romney

```python
from sklearn.model_selection import train_test_split
X = df_votes[['White','age65plus']]
y = df_votes['16Winner']
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=42)
```

```python
# from sklearn.svm import SVC  
# svclassifier = SVC(kernel='poly', degree=8)  
# svclassifier.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC  
svclassifier = SVC(kernel='rbf')  
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)  
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
plot_decision_regions(X=X.values, 
                      y=y.values,
                      clf=svclassifier, 
                      legend=2)
plt.xlabel(X.columns[0], size=14)
plt.ylabel(X.columns[1], size=14)
plt.title('SVM Decision Region Boundary', size=16)
```

```python
df_votes['turnout_diff_disc'] = df_votes.apply(lambda row: 1 if (row['turnout_diff'] > 0) else 0 ,axis=1)
```

```python
df_votes
```

```python
le = LabelEncoder()
le.fit(df_votes['turnout_diff_disc'])
df_votes['turnout_diff_disc'] = le.transform(df_votes['turnout_diff_disc'])
```

```python
X = df_votes[['percent_diff', 'NonEnglish']]
y = df_votes['turnout_diff_disc']
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=42)
svclassifier = SVC(kernel='rbf') 
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)  
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```

```python
plot_decision_regions(X=X.values, 
                      y=y.values,
                      clf=svclassifier, 
                      legend=2)
plt.xlabel(X.columns[0], size=14)
plt.ylabel(X.columns[1], size=14)
plt.title('SVM Decision Region Boundary', size=16)
```

```python
from sklearn.tree import DecisionTreeClassifier  
X = df_votes[['percent_diff', 'White', 'Income','NonEnglish','age65plus','Poverty','12Winner','Density','population2014']]
y = df_votes['turnout_diff_disc']
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=42)
classifier = DecisionTreeClassifier()  
classifier.fit(X_train, y_train)  
```

```python
y_pred = classifier.predict(X_test)  
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```

```python
classifier.feature_importances_
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df_votes[['percent_diff', 'White', 'Income','NonEnglish','age65plus','Poverty','12Winner','Density','population2014']]
y = df_votes['turnout_diff']
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=42)
regressor = DecisionTreeRegressor()  
regressor.fit(X_train, y_train)
```

```python
regressor.feature_importances_
```

```python
y_pred = regressor.predict(X_test)
df_regress=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})  
df_regress 
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  
```

```python
ax1 = df_regress.plot.scatter(x='Actual', y='Predicted', figsize=(10,10))
```
