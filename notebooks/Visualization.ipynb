```python
df_votes = pd.read_csv('../data/votes.csv')
df_votes.head(10)
```

```python
for i,row in enumerate(df_votes.iterrows()):
    if row[1].votes_gop_2016>row[1].votes_dem_2016:
        df_votes.loc[i,'16Winner'] = 'Trump'
    else:
        df_votes.loc[i,'16Winner'] = 'Clinton'  
    if row[1].votes_gop_2012>row[1].votes_dem_2012:
        df_votes.loc[i,'12Winner'] = 'Romney'
    else:
        df_votes.loc[i,'12Winner'] = 'Obama' 
        
df_votes.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df_votes['16Winner'])
df_votes['16Winner'] = le.transform(df_votes['16Winner'])
le.fit(df_votes['12Winner'])
df_votes['12Winner'] = le.transform(df_votes['12Winner'])

df_votes.head(10)
```
